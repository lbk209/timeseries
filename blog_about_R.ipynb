```r
# auto.arima, forcast, AirPassengers를 가진 forecast 패키지 로드
library(forecast)

# AirPassengers 데이터를 ARIMA로 모델링 (ARIMA 모델 파라미터 최적화 자동 수행)
fit <- auto.arima(AirPassengers)

# 완성된 ARIMA 모델로 향후 12개월치 예측
y <- forecast(fit, h=12)

# 과거값과 예측값 그래프 출력
plot(y)
```

```r
data.seq = seq(1,6)
data.seq
```

```r
data.ts = ts(data.seq)
data.ts
```

```r
library(timeSeries)
data.tS <- timeSeries(data.seq)
data.tS
```

```r
data.matrix = matrix(data.ts, 3, 2)
data.matrix
```

```r
data.ts.2 = ts(data.matrix)
data.ts.2
```

```r
library(xts) 
order.by <- seq(as.Date('2020-01-01'), by=1, length.out=nrow(data.ts.2))
data.xts <- xts(data.ts.2, order.by=order.by)
data.xts
```

```r
data.zoo <- zoo(data.ts.2, order.by)
data.zoo
```

```r
data.df <- as.data.frame(data.ts.2)
data.df
```
